```julia
using PyPlot
using Distributions
```

```julia
function GeoBM(mu, sigma, S0, ctime, etime, M, N)
    tau = etime - ctime;
    Deltat = tau/M
    Deltax = sqrt(Deltat)
    k = 0:1:M
    tk = ctime + k*Deltat
    
    srand(34567)
    
    for i = 1:1:N
        W = 0;
        xrand = (2*rand(Binomial(1,0.5), M+1, 1)-1)*Deltax
        W = cumsum(xrand)
        S = S0 .* exp.( (mu-sigma^2/2).*(tk-ctime) + sigma*W)
        plot(tk, S)
    end
    
    title("Geometric brownian motion")
    xlabel("time u")
    ylabel(L"$S_u$")
end
```

```julia
GeoBM(0.06, 0.09, 100, 0, 1, 128, 50)
```
